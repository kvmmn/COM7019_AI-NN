```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# Set a manual seed for reproducibility
torch.manual_seed(274)

# Generate synthetic dataset
num_samples = 1000
x = torch.randn(num_samples, 2) # 1000 samples, 2 features
true_weights = torch.tensor([3, 5], dtype=torch.float32) # True weights for the linear model
true_bias = torch.tensor([3], dtype=torch.float32) # True bias for the linear model
y = x @ true_weights + true_bias # Linear relationship: y = 3*x1 + 5*x2 + 3

# Visualize the dataset
fig, ax = plt.subplots(1, 2, sharey=True)
ax[0].scatter(x[:, 0], y)
ax[1].scatter(x[:, 1], y)
ax[0].set_xlabel('x1'); ax[0].set_ylabel('Y')
ax[1].set_xlabel('x2'); ax[1].set_ylabel('Y')
plt.suptitle('Scatter Plots of the Dataset')
plt.show()
```
